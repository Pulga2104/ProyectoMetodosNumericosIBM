1. Utilice aritmética de corte de tres dígitos para calcular las siguientes sumas. Para cada parte, ¿qué método es más preciso y por qué?

    a. $\Sigma_{i=1}^{10}(\frac{1}{i^{2}})$ primero por $\frac{1}{1}+\frac{1}{4}+\cdot\cdot\cdot+\frac{1}{100}$ y luego por $\frac{1}{100}+\frac{1}{81}+\cdot\cdot\cdot+\frac{1}{1}$ 4

    Pseudocódigo:

    Inicio

            suma1 ← 0

            suma2 ← 0

            Para i ← 1 Hasta 10 Hacer

                suma1 ← suma1 + truncar(1 / (i*i), 3)

            FinPara

            Para i ← 10 Hasta 1 Paso -1 Hacer

                suma2 ← suma2 + truncar(1 / (i*i), 3)

            FinPara

            Imprimir "Suma directa:", suma1

            Imprimir "Suma inversa:", suma2
        
    Fin



```python
def truncar(num, digitos):
    return float(str(num)[:str(num).find('.') + digitos + 1])

suma1 = 0
suma2 = 0

# Suma directa: del 1 al 10
for i in range(1, 11):
    suma1 += truncar(1 / (i**2), 3)

# Suma inversa: del 10 al 1
for i in range(10, 0, -1):
    suma2 += truncar(1 / (i**2), 3)

print("Suma directa:", suma1)
print("Suma inversa:", suma2)
```

Sumar en orden inverso suele ser más preciso porque los números pequeños se agregan al final.
