1. Use el método de bisección para encontrar soluciones precisas dentro de $10^{-2}$ para $x^3 - 7x^2 + 14x - 6 = 0$ en cada intervalo.
a. $[0, 1]$
b. $[1, 3.2]$
c. $[3.2, 4]$


```python
import math
from scipy.optimize import bisect

def f1(x):
    return x**3 - 7*x**2 + 14*x - 6

def ejercicio1():
    print("Ejercicio 1: f(x) = x^3 - 7x^2 + 14x - 6")

    intervalos = [(0, 1), (1, 3.2), (3.2, 4)]
    for a, b in intervalos:
        raiz = bisect(f1, a, b, xtol=1e-2)
        print(f"  Intervalo [{a}, {b}] -> raíz = {raiz:.6f}")
        
ejercicio1()
```

5. Sea $f(x) = (x + 3)(x + 1)^2 x (x - 1)^3 (x - 3)$. ¿En qué cero de $f$ converge el método de bisección cuando se aplica en los siguientes intervalos?
a. $[-1.5, 2.5]$
b. $[-0.5, 2.4]$
c. $[-0.5, 3]$
d. $[-3, -0.5]$

```python
import math
from scipy.optimize import bisect

def f5(x):
    return (x + 3)*(x + 1)**2 * x * (x - 1)**3 * (x - 3)

def ejercicio5():
    print("\nEjercicio 5: función con múltiples ceros")
    intervalos = [
        (-1.5, 2.5),
        (-0.5, 2.4),
        (-0.5, 3),
        (-3, -0.5)
    ]

    for a, b in intervalos:
        print(f"Intervalo [{a}, {b}]")
        try:
            raiz = bisect(f5, a, b, xtol=1e-6)
            print(f"Converge a raíz ≈ {raiz:.6f}")
        except ValueError:
            print("Bisección inválida")
            
ejercicio5()
```

1. Un abrevadero de longitud $L$ tiene una sección transversal en forma de semicírculo con radio $r$. Cuando se llena con agua hasta una distancia $h$ a partir de la parte superior, el volumen $V$ de agua es$$V = L \left[ 0.5\pi r^2 - r^2 \arcsin(h/r) - h(r^2 - h^2)^{1/2} \right]$$Suponga que $L = 10 \text{ cm}$, $r = 1 \text{ cm}$ y $V = 12.4 \text{ cm}^3$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 \text{ cm}$.

```python
import math
from scipy.optimize import bisect

def volumen(h, L=10, r=1):
    return L * (0.5*math.pi*r**2 - r**2*math.asin(h/r) - h*math.sqrt(r**2 - h**2))

def g1(h):
    return volumen(h) - 12.4

def aplicado1():
    print("\nAplicado 1: abrevadero")
    raiz = bisect(g1, 0.1, 0.2, xtol=1e-6)
    profundidad = 2 - raiz
    print(f"h desde arriba ≈ {raiz:.4f} cm")
    print(f"Profundidad del agua ≈ {profundidad:.4f} cm")
    
aplicado1()
```

2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa 1$m$ cae desde una altura 2$s_0$ y que la altura del objeto después de 3$t$ segundos es4$$s(t) = s_0 - \frac{mg}{k}t + \frac{m^2g}{k^2} \left( 1 - e^{-kt/m} \right),$$donde $g = 9.81 \text{ m/s}^2$ y $k$ representa el coeficiente de la resistencia del aire en $\text{Ns/m}$. Suponga $s_0 = 300 \text{ m}$, $m = 0.25 \text{ kg}$ y $k = 0.1 \text{ Ns/m}$. Encuentre, dentro de $0.01 \text{ segundos}$, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import math
from scipy.optimize import bisect

def s(t, s0=300, m=0.25, k=0.1, g=9.81):
    return s0 - (m*g/k)*t + (m**2 * g / k**2)*(1 - math.e**(-k*t/m))

def aplicado2():
    print("\nAplicado 2: caída con resistencia")
    raiz = bisect(s, 10, 20, xtol=0.01)
    print(f"Tiempo de caída ≈ {raiz:.2f} s")
    
aplicado2()
```
